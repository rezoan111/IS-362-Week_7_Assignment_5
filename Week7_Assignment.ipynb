```python
import pandas as pd

# Load the movie ratings file
movies = pd.read_csv("movie_ratings.csv")

# Show the data
movies
```

```python
# Average rating by user
# In this part, I am finding the average movie rating for each user.

user_avg = movies.set_index("User").mean(axis=1)

user_avg
```

```python
# Average rating by movie
# In this part, I am finding the average rating for each movie.

movie_avg = movies.drop("User", axis=1).mean(axis=0)

movie_avg
```

```python
# Normalized ratings by user
# In this part, I am normalizing the ratings by subtracting
# each user's average rating from that user's movie ratings.

ratings_only = movies.set_index("User")

normalized = ratings_only.sub(user_avg, axis=0)

normalized
```

```python
# Average normalized rating by user
# In this part, I am finding the average normalized rating for each user.

normalized_user_avg = normalized.mean(axis=1)

normalized_user_avg
```

```python
# Average normalized rating by movie
# In this part, I am finding the average normalized rating for each movie.

normalized_movie_avg = normalized.mean(axis=0)

normalized_movie_avg
```

# Conclusion

In this assignment, I created a movie ratings DataFrame for 5 users and 6 movies. Then I calculated the average rating by user and the average rating by movie. Based on the original averages, **American Sniper** had the highest average rating at **4.67**, while **Jurassic World** had the lowest average rating at **2.50**.

Next, I normalized the ratings by subtracting each user's average rating from that user's movie ratings. After normalization, the average normalized rating for each user was about **0**, which is expected. This shows that the normalization worked correctly.

The normalized movie averages showed that **American Sniper** still had the highest adjusted rating at about **1.06**, while **Jurassic World** had the lowest adjusted rating at about **-0.92**. This means American Sniper was rated above users’ personal averages, while Jurassic World was rated below users’ personal averages.

One advantage of normalized ratings is that they make movie comparisons more fair because they adjust for users who rate movies higher or lower than other users. This helps reduce user bias in the ratings. One disadvantage is that normalized ratings can be harder to understand than the original ratings, because the values are no longer the actual 1 to 5 movie scores. Another disadvantage is that normalization may remove some information about the users’ true rating style.

Overall, normalized ratings are useful when comparing movies across different users, but the original ratings are still easier to read and understand.
